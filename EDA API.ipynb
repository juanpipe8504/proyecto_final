Gráfica de descripción del clima

```python
variables_interes = ["clear sky", "broken clouds", "scattered clouds", "overcast clouds", "few clouds", "light rain", "moderate rain"]
filtered_df = df[df['Descripción clima'].isin(variables_interes)]
descripcion_clima_counts = filtered_df['Descripción clima'].value_counts()

colors = ['blue', 'red', 'green', 'yellow', 'purple', 'pink', 'gray']

plt.figure(figsize=(8, 8))
plt.pie(descripcion_clima_counts, labels=descripcion_clima_counts.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Descripción de Clima')
plt.axis('equal')
plt.show()
```

```python
variables_interes = ['fog', 'mist', 'light intensity drizzle', 'haze', 'shower rain', 'thunderstorm with heavy rain']
filtered_df = df[df['Descripción clima'].isin(variables_interes)]
descripcion_clima_counts = filtered_df['Descripción clima'].value_counts()

colors = ['blue', 'red', 'green', 'yellow', 'purple', 'pink', 'gray']

plt.figure(figsize=(8, 8))
plt.pie(descripcion_clima_counts, labels=descripcion_clima_counts.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Descripción de Clima')
plt.axis('equal')
plt.show()
```

```python
variables_interes = ['thunderstorm with light rain', 'smoke', 'thunderstorm', 'heavy intensity rain', 'heavy intensity shower rain']
filtered_df = df[df['Descripción clima'].isin(variables_interes)]
descripcion_clima_counts = filtered_df['Descripción clima'].value_counts()

colors = ['blue', 'red', 'green', 'yellow', 'purple']

plt.figure(figsize=(8, 8))
plt.pie(descripcion_clima_counts, labels=descripcion_clima_counts.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Descripción de Clima')
plt.axis('equal')
```
