```python
import requests
import pandas as pd

def extract():
    # URL para obtener datos del clima en Londres
    url = 'http://api.openweathermap.org/data/2.5/weather?q=London&appid=25cb5545c16d4f98a0578df496228c74'

    try:
        response = requests.get(url)
        response.raise_for_status()  # Verificar si la solicitud fue exitosa

        # Obtener datos del clima en formato JSON
        data = response.json()
        weather_data = {
            'description': data['weather'][0]['description'],
            'icon': data['weather'][0]['icon'],
            'id': data['weather'][0]['id'],
            'main': data['weather'][0]['main']
        }
        print('Weather data:', weather_data)

    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return

    # Crear DataFrame con los datos del clima
    df = pd.DataFrame({'weather': [weather_data]})
    df.to_csv('data.csv', index=False)  # Guardar en un archivo CSV

if __name__ == "__main__":
    extract()



```

```python
from datetime import timedelta
from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime
from etl import extract, transform, load

default_args = {
    'owner': 'airflow',
    'depends_on_past': False,
    'start_date': datetime(2023, 9, 13),  # Update the start date to today or an appropriate date
    'email': ['airflow@example.com'],
    'email_on_failure': False,
    'email_on_retry': False,
    'retries': 1,
    'retry_delay': timedelta(minutes=1)
}

with DAG(
    'api_dag',
    default_args=default_args,
    description='Our first DAG with ETL process!',
    schedule_interval='@daily',  # Set the schedule interval as per your requirements
) as dag:

    extract_task = PythonOperator(
        task_id='extract_task',
        python_callable=extract,
    )

    transform_task = PythonOperator(
        task_id='transform_task',
        python_callable=transform,
        )

    load_task = PythonOperator(
        task_id='load_task',
        python_callable=load,
        )

    extract_task >> transform_task >> load_task
```
